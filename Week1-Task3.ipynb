```python
pivot = pd.pivot_table( df,
    index="shopping_mall",
    columns="category",
    values="price",
    aggfunc="sum")

plt.figure(figsize=(12,6))

plt.imshow(pivot, aspect="auto")

plt.colorbar(label="Revenue")

plt.xticks(range(len(pivot.columns)), pivot.columns, rotation=90)

plt.yticks(range(len(pivot.index)), pivot.index)

plt.title("Revenue Heatmap: Shopping Mall vs Category")

plt.xlabel("Category")

plt.ylabel("Shopping Mall")

plt.show()
```

### Bu heatmap hər ticarət mərkəzində hansı məhsul kateqoriyasının daha çox gəlir gətirdiyini göstərir. Daha tünd rənglər daha yüksək gəliri ifadə edir.

```python
plt.figure(figsize=(12,6))
for column in df.groupby(["Year", "category"])["price"].sum().unstack().columns:
    plt.plot(df.groupby(["Year", "category"])["price"].sum().unstack().index, df.groupby(["Year", "category"])["price"].sum().unstack()[column], marker="o", label=column)
plt.title("Yearly Revenue Trend by Category")
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.legend()
plt.show()
```

### Bu qrafik 2021–2023-cü illərdə hər məhsul kateqoriyasının gəlir dəyişimini göstərir.

```python
df.groupby(["category", "gender"])["price"].sum().unstack().plot(kind="bar", color='#AB7020',edgecolor='#36250F')
plt.title("Male vs Female Spending by Category")
plt.xlabel("Category")
plt.ylabel("Total Spending")
plt.show()
```
